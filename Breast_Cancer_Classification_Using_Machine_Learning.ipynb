# Breast_Canser_classification_Using_Machine_Learning

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection & Processing


```python
# Loading the data from  sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# Loading the data to a dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data , columns = breast_cancer_dataset.feature_names)
```

```python
# Print first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the target column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# Print the last 5 rows of the dataframe
data_frame.tail()
```

```python
# Number of the rows and columns in the dataset
data_frame.shape
```

```python
data_frame.info()
```

```python
data_frame.isnull().sum()
```

```python
# Statistical measures about the data
data_frame.describe()
```

```python
# Checking the distribution of Target Varibale
data_frame['label'].value_counts()
```

## 1 -> Benign
## 0 -> Malignant

```python
data_frame.groupby('label').mean()
```

Separating the features and the targets

```python
X = data_frame.drop(columns = 'label' , axis = 1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data & Testing data

```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.2 , random_state= 2)
```

```python
X_train
```

```python
print(X.shape , X_train.shape , X_test.shape)
```

## Model Training

## Logistic Regression

```python
model = LogisticRegression()
```

```python
# Training the logistic Regression model using Training data

model.fit(X_train , Y_train )
```

```python
data_frame.shape
```

```python
data_frame.info()
```

Model Evaluation

## Accuracy score

```python
# Accuracy on trainig data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train , X_train_prediction)
```

```python
print('Accuracy on training data = ' , training_data_accuracy)
```

```python
# accuracy on the test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test , X_test_prediction)
```

```python
print("Accuracy on test data = " , test_data_accuracy)
```

## Building a Predictive System


```python
input_data = (11.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,0.2433,0.06563)

# Change the target input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1 , -1)

# Standardizing the input data
input_data_std = input_data_reshaped

prediction = model.predict(input_data_std)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
  print('The tumor is Malignant')

else:
  print('The tumor is Benign')
```
